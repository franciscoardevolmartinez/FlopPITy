# ARCiS Retrieval

This notebook shows the minimal setup for a retrieval with ARCiS. 

```python
from floppity import Retrieval
from floppity.simulators import ARCiS, read_ARCiS_input
```

## Paths

Edit these paths for your local ARCiS setup.

```python
arcis_input = "path/to/arcis_input.in"
arcis_executable = "/usr/local/bin/ARCiS" # not needed if this is the path

arcis_output_dir = "output_ARCiS_multi_component"
```

## Read The Single-Component ARCiS Setup

`read_ARCiS_input` returns the fitted parameters and observation files from one ARCiS input file. For a binary retrieval, the helper below duplicates component-specific parameters and keeps selected parameters shared.

## Build The Retrieval

The retrieval uses the wrapped simulator returned by `make_binary_simulator`.

```python
pars, obs_dict = read_ARCiS_input(arcis_input)
R = Retrieval(ARCiS)
R.parameters = pars
R.get_obs(obs_dict)
```

## Runtime Options

Start small while testing that ARCiS runs and the parameter ordering is correct. Increase `n_samples`, `n_samples_init`, and `n_rounds` for a real retrieval.

```python
ARCiS_kwargs = {
    "input_file": "path/to/arcis_input.in",
    "output_dir": "path/to/arcis/output",
    "ARCiS_dir": "path/to/arcis/executable", #can be omitted if it's /usr/local/bin/ARCiS (by default)
}
```

## Run

```python
R.run(
    simulator_kwargs=ARCiS_kwargs,
    output_dir=ARCiS_kwargs['output_dir'],
    n_threads=2, #typically as many as possible
    save_data=True,
)
```

## Inspect The Result

```python
fig = R.plot_corner(n_samples=1000)
```
