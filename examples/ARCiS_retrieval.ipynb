# ARCiS Retrieval

This notebook shows the minimal setup for a retrieval with ARCiS. 

```python
from floppity import Retrieval
from floppity.simulators import ARCiS, read_ARCiS_input
```

## Read The ARCiS input file

`read_ARCiS_input` returns the fitted parameters and observation files from one ARCiS input file. 

```python
ARCiS_kwargs = {
    "input_file": "path/to/arcis_input.in",
    "output_dir": "path/to/arcis/output",
    "ARCiS_dir": "path/to/arcis/executable", #can be omitted if it's /usr/local/bin/ARCiS (by default)
}

pars, obs_dict = read_ARCiS_input(ARCiS_kwargs['input_file'])
```

## Build The Retrieval

```python
R = Retrieval(ARCiS)
R.parameters = pars
R.get_obs(obs_dict)
```

## Run

```python
R.run(
    simulator_kwargs=ARCiS_kwargs,
    output_dir=ARCiS_kwargs['output_dir'],
    n_threads=2, #typically as many as possible
    save_data=True,
)
```

## Inspect The Result

```python
fig = R.plot_corner(n_samples=1000)
```
